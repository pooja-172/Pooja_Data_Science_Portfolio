# Exploratory Data Analysis on COVID-19 World Vaccination Progress

Coronavirus disease 2019 (COVID-19) is a contagious disease caused by a virus, the severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2). The first known case was identified in Wuhan, China, in December 2019. The disease quickly spread worldwide, resulting in the COVID-19 pandemic. Symptoms of COVID‑19 are variable, but often include fever, cough, headache, fatigue, breathing difficulties, loss of smell, and loss of taste. Symptoms may begin one to fourteen days after exposure to the virus. At least a third of people who are infected do not develop noticeable symptoms. 

Preventive measures to reduce the chances of infection include getting vaccinated, staying at home, wearing a mask in public, avoiding crowded places, keeping distance from others, ventilating indoor spaces, managing potential exposure durations, washing hands with soap and water often and for at least twenty seconds, practising good respiratory hygiene, and avoiding touching the eyes, nose, or mouth with unwashed hands.

The first COVID‑19 vaccine was granted regulatory approval on 2 December 2020 by the UK medicines regulator MHRA. It was evaluated for emergency use authorization (EUA) status by the US FDA, and in several other countries.

```python
df = df.dropna()
df.head()
```

#### Total vaccinations with respect to country

```python
plt.subplots(figsize=(20,5))
df.groupby('country')['total_vaccinations'].mean().sort_values(ascending=False)[:60].plot(kind='bar',grid=True,fontsize=15,colormap='Accent')
plt.title('Total Vaccinations Per Country',fontsize=20)
plt.ylabel('Total vaccinations',fontsize=17)
plt.xlabel('Country',fontsize=17)
```

#### Number of people partially vaccinated per hundred with respect to country

```python
plt.figure(figsize=(20,5))
df.groupby('country')['people_vaccinated_per_hundred'].mean().sort_values(ascending=False)[:60].plot(kind='bar',grid=True,fontsize=15,colormap='rainbow')
plt.title('Number of people partially vaccinated per hundred in each country',fontsize=20)
plt.xlabel('country',fontsize=17)
plt.ylabel('Number of people vaccinated',fontsize=17)
```

#### Number of people fully vaccinated per hundred with respect to country

```python
plt.figure(figsize=(20,5))
df.groupby('country')['people_fully_vaccinated_per_hundred'].mean().sort_values(ascending=False)[:60].plot(kind='bar',fontsize=15,grid=True,colormap='Set1')
plt.title('Number of people fully vaccinated per hundred in each country',fontsize=20)
plt.xlabel('country',fontsize=17)
plt.ylabel('Number of people vaccinated',fontsize=17)
```

```python
sns.pairplot(data=df)
```

#### Top 5 fully vaccinated countries

```python
plt.figure(figsize=(10,5))
df.groupby('country')['people_fully_vaccinated_per_hundred'].max().sort_values(ascending=False)[:5].plot(kind='bar',fontsize=15,colormap='Set3')
plt.title('Top 5 Fully Vaccinated Countries',fontsize=15)
plt.xlabel('country',fontsize=15)
plt.ylabel('mean vaccinated population',fontsize=15)
```

```python
plt.figure(figsize=(12,9))
cor = df.corr()
sns.heatmap(cor,annot=True,cmap='crest_r')
```

```python
india = df[df['country']=='India']
india= india.set_index(india['date'])
india.head()
```

#### Comparison between the number of partially vaccinated people and fully vaccinated people 

```python
india[['people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred']].plot.line(figsize=(17,4),fontsize=15)
plt.title('Comparison between the number of partially vaccinated people and fully vaccinated people in India per hundred',fontsize=20)
plt.xlabel('Date',fontsize=20)
plt.ylabel('Vacc ppl per 100',fontsize=20)
plt.legend(fontsize=15,facecolor='lightgrey')
```

#### Vaccine(s) used in each country

```python
vaccine = df[['country','vaccines']]
new = vaccine.drop_duplicates()
new.reset_index()
```

#### Extracting the month and year from date column 

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda date : date.year)
```

```python
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
df.head()
```

#### Relationship between year and total number of people who are fully vaccinated (directly proportional)

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='year',y='people_fully_vaccinated_per_hundred',data=df,color='red')
```

```python
vacc = df.groupby(by='vaccines').sum().sort_values(by='daily_vaccinations',ascending=False)[0:20]
```

```python
vacc
```

#### Top 10 Vaccines Distributed by Total Vaccination

```python
plt.figure(figsize=(12,10))
sns.barplot(data=vacc,x=vacc.index,y='daily_vaccinations',palette='nipy_spectral')
plt.xticks(rotation='vertical')
plt.title("Top 10 Vaccines Distributed by Total Vaccination",fontsize=17)
plt.ylabel("daily_vaccinations",fontsize=15)
plt.xlabel("vaccines",fontsize=15)
```
