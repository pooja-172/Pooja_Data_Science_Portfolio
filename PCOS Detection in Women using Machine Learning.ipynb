### PCOS Detection in Women using Machine Learning 

Polycystic ovary syndrome (PCOS) is a common female endocrine disorder that affects 6-15% of the female population. Many women have PCOS but don’t know it. In one study, up to 70 percent of women with PCOS hadn’t been diagnosed. PCOS affects up to almost 27 percent of women during their childbearing years. It involves cysts in the ovaries, high levels of male hormones, and irregular periods.

Using machine learning, to detect whether a woman is suffering from PCOS, different machine learning algorithms/classifiers can be used like Random Forest classifier, Gaussian Naive Bayes, Support Vector Machines, Logistic Regression etc,.
In this project, I have used Logistic Regression and Random Forest classifier for the detection of the disorder.

### 4. Building the model 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
X = pcos[[' Age (yrs)', 'Weight (Kg)','Height(Cm) ','BMI', 'Blood Group', 'Pulse rate(bpm) ',
             'RR (breaths/min)', 'Hb(g/dl)', 'Cycle(R/I)', 'Cycle length(days)',
             'Marraige Status (Yrs)', 'Pregnant(Y/N)', 'No. of aborptions',
             '  I   beta-HCG(mIU/mL)', 'FSH(mIU/mL)',
             'LH(mIU/mL)', 'FSH/LH', 'Hip(inch)', 'Waist(inch)', 'Waist:Hip Ratio',
             'TSH (mIU/L)', 'PRL(ng/mL)', 'Vit D3 (ng/mL)',
             'PRG(ng/mL)', 'RBS(mg/dl)', 'Weight gain(Y/N)', 'hair growth(Y/N)',
             'Skin darkening (Y/N)', 'Hair loss(Y/N)', 'Pimples(Y/N)',
             'Fast food (Y/N)', 'Reg.Exercise(Y/N)', 'BP _Systolic (mmHg)',
             'BP _Diastolic (mmHg)', 'Follicle No. (L)', 'Follicle No. (R)',
             'Avg. F size (L) (mm)', 'Avg. F size (R) (mm)', 'Endometrium (mm)']]
y = pcos['PCOS (Y/N)']
```

Performing train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### 4.1 Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model_predictions = model.predict(X_test)
```

```python
print('Classification Report : ',classification_report(y_test,model_predictions))
print('Confusion Matrix: ',confusion_matrix(y_test,model_predictions))
```

```python
log_acc = accuracy_score(y_test,model_predictions)
print('Accuracy of the logistic regression model: {}{}'.format(round(log_acc*100,2)," %"))
```

Thus we can conclude that our model performed at an accuracy of 82.58%

### 4.2 Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand_pcos = RandomForestClassifier()
```

```python
rand_pcos.fit(X_train,y_train)
```

```python
rand_pred = rand_pcos.predict(X_test)
```

```python
print('Classification Report: \n',classification_report(y_test,rand_pred))
```

```python
print('Confusion Matrix: \n',confusion_matrix(y_test,rand_pred))
```

```python
random_acc = accuracy_score(y_test,rand_pred)
print('Accuracy of the model built using Random Forest Classifier: {}{}'.format(round(random_acc*100,2)," %"))
```

Thus we can conclude that our model performed at an accuracy of 88.76%

Thus we have used logistic regression and random forest classifier to find the accuracy levels separately. For our model, the accuracy score of the random forest classifier model is greater than that of the logistic regression model.
